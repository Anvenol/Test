```python
import pandas as pd
```

```python
data = pd.read_csv('D:/Study/深度学习/data.csv')
```

```python
data
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=data.Education
y=data.Income
```

```python
plt.scatter(x,y)
```

```python
model=tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1,input_shape=(1,)))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
                loss='mse'
)
```

```python
history=model.fit(x,y,epochs=5000)
```

```python
model.predict(x)
```

```python
model.predict(pd.Series([30]))
```
